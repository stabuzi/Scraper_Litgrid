```python
import requests
from bs4 import BeautifulSoup
from typing import List, Dict, Optional
import pandas as pd
```

```python
def get_pages_info(URL1: str) -> Optional[List[Dict]]:

    results: List[Dict] = []
    results1: List[Dict] = []
    Find_all_TS: bool

    results.append({
                "Failo URL": "", 
                "TS pavadiniams": "", 
                "Kategorijos": "", 
                "Puslapio URL": Main_URL
    })   
    Find_all_TS=False 
    i=1  
    while Find_all_TS == False:
            Find_all_TS=True  
            for item in results:

                if item["Failo URL"]=="":
                    page=requests.get(item["Puslapio URL"])
                    #from_encoding="iso-8859-1"
                    soup = BeautifulSoup(page.content, 'html.parser')
                    j=1
                    try:
                        SR=soup.find("div", class_="inner_bg_top").find_all("a",class_="file")
                    except:
                        j=0
                        SR=False
                    if SR:
                        for Elementas in SR:
                            results1.append({
                            "Failo URL": Elementas["href"], 
                            "TS pavadiniams": Elementas.text.strip(), 
                            "Kategorijos": item["Kategorijos"], 
                            "Puslapio URL": item["Puslapio URL"]
                            })

                    elif j==1:
                        SR=soup.find("div", class_="inner_bg_top").find_all("a")
                        for Elementas in SR:
                            Find_all_TS=False
                            results1.append({
                            "Failo URL": "", 
                            "TS pavadiniams": '"', 
                            "Kategorijos":item["Kategorijos"] + "/" + Elementas.text, 
                            "Puslapio URL": Elementas["href"]
                            })
                else:
                    results1.append({
                    "Failo URL": item["Failo URL"], 
                    "TS pavadiniams": item["TS pavadiniams"], 
                    "Kategorijos": item["Kategorijos"], 
                    "Puslapio URL": item["Puslapio URL"]
                    })
            results=results1.copy()
            results1.clear()
            i=i+1
    return results
```

```python

Main_URL="https://www.litgrid.eu/index.php/tinklo-pletra/standartiniai-techniniai-reikalavimai/standartiniai-techniniai-reikalavimai/2632"
```

```python
results = get_pages_info(Main_URL)
results_df = pd.DataFrame(results)
results_df.head()
results_df.to_csv("litgrid.csv")
```
